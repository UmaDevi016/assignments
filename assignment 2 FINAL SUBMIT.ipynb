```python
import numpy as np
values = [-2000,-1000,0,1000,2000,3000]
probabilities = [0.1,0.1,0.2,0.2,0.3,0.1]
expected_return = 800
```

```python
variance = sum([(x - expected_return) ** 2 * p for x, p in zip(values, probabilities)])
```

```python
standard_deviation = np.sqrt(variance)
```
